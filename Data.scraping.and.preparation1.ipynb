Step 1: Scrape your competitor's data 
-----------------
We scraped data using BeautifulSoup, searching by class
then by table then add all data into lists then put it in dataframe using pandas:

```python

from bs4 import BeautifulSoup
import requests
import pandas as pd

source = requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares').text
soup = BeautifulSoup(source, 'html')
header =soup.find('table', class_ = 'table table-striped table-responsive-md')
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
for row in header.find_all('tr')[1:51]:
    cell = row.find_all('td')
    a.append(cell[0].text)
    b.append(cell[1].text)
    c.append(cell[2].text)
    d.append(cell[3].text)
    e.append(cell[4].text)
    f.append(cell[5].text)
    g.append(cell[6].text)
    h.append(cell[7].text)
df = pd.DataFrame({'rank':a, 'x_calss':b, 'date':c, 'region':d, 'start_time':e, 'max_time':f, 'end_time':g, 'movie':h})
df    

 
```
